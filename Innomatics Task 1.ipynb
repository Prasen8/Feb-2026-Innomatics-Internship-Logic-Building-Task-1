```python
"""
Data Science Internship – February 2026
Logic Building Task – 1

Author: <Your Name>
Description:
This file contains solutions for basic logic building tasks including:
1. User Login Check
2. Pass / Fail Analyzer
3. Simple Data Cleaner
4. Message Length Analyzer
5. Error Message Detector
"""


# 1️⃣ User Login Check:

print("----- User Login Check -----")

username = "admin"
password = "1234"

input_username = "admin"      # Change value to test
input_password = "1234"       # Change value to test

if input_username == username and input_password == password:
    print("Login Successful")
else:
    print("Invalid Credentials")

print()



# 2️⃣ Pass / Fail Analyzer :

print("----- Pass / Fail Analyzer -----")

marks = [45, 78, 90, 33, 60]

pass_count = 0
fail_count = 0

for mark in marks:
    if mark >= 50:
        pass_count += 1
    else:
        fail_count += 1

print("Total Students:", len(marks))
print("Passed Students:", pass_count)
print("Failed Students:", fail_count)

print()



# 3️⃣ Simple Data Cleaner :

print("----- Simple Data Cleaner -----")

names = [" Alice ", "bob", " CHARLIE "]

cleaned_names = []

for name in names:
    clean_name = name.strip().lower()
    cleaned_names.append(clean_name)

print("Original Names:", names)
print("Cleaned Names:", cleaned_names)

print()



# 4️⃣ Message Length Analyzer :

print("----- Message Length Analyzer -----")

messages = ["Hi", "Welcome to the platform", "OK"]

for message in messages:
    length = len(message)
    print(f"Message: '{message}' | Length: {length}")
    
    if length > 10:
        print("⚠️  This message is longer than 10 characters.")

print()



# 5️⃣ Error Message Detector :

print("----- Error Message Detector -----")

logs = ["INFO", "ERROR", "WARNING", "ERROR"]

error_count = logs.count("ERROR")

print("Total ERROR entries:", error_count)

print("\n✅ All Tasks Completed Successfully!")
```
